```python
a=pd.read_csv('directors1.csv')
a
```

```python
b=pd.read_csv('moviesdata.csv')
b
```

```python
df=a.merge(b,left_on='id',right_on='director_id',how='right')
df
```

```python
df.drop(columns=['Unnamed: 0_x','Unnamed: 0_y','id_x','id_y'],inplace=True)
df
```

```python
df.info()
```

```python
df.duplicated()
```

```python
df.isna()
```

```python
df.sort_values(by='budget',ascending=False)
df.head(15)
```

```python
df.groupby(['director_id','director_name','title','budget'])['popularity'].max().to_frame()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df['year'],df['budget'],marker='*')
plt.title('the trends in budget')
plt.xlabel('year')
plt.ylabel('budget')
plt.legend(['year','budget'])
plt.show()
```

```python
df
```

```python
df.groupby(['title'])['popularity'].max().sort_values(ascending=False).head(10).plot(kind='pie',autopct='%1.1f%%');
plt.title('All Time Popular Movies');
```

* The top 10 popular movies are listed above with 'Interstellar' on the first position with 27.8% and Fight Club with 5.6% at the last position

```python
df['day'].value_counts().plot(kind='bar');
```

* Most of the films were released on Friday with the least releases on saturdays

```python
df['year'].value_counts().plot(kind='bar')
plt.title('Release trends');
```

* 2006 is plotted with highest no.of movie release against 1976 with the least no.of releases

```python
df.groupby('director_name')['revenue'].sum().sort_values(ascending=False).head(10).plot(kind='barh');
plt.title('Directors and Revenue')
plt.xlabel('revenue')
```

* the top 10 directors are plotted based on their revenue generation where Steven Spielberg's movies has generated more revenue in comparison to other directors.

```python
a=df.sort_values(by='revenue',ascending=False)
data=a.head(10)
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].bar(data['title'], data['revenue'])
axes[0].set_title('Revenue Generated')
axes[0].set_xlabel('Title')
axes[0].tick_params(axis='x', rotation=90) 


axes[1].bar(data['title'], data['budget'])
axes[1].set_title('Profit Generated')
axes[1].set_xlabel('Title')
axes[1].tick_params(axis='x', rotation=90) 

plt.tight_layout()
plt.show()
```

* the top 10 revenue generated movies are compared with their budget. results shows that among the top 10 revenue generated films Pirates of the Caribbean: On Stranger Tides is the one with comparitively higher budget against less income and Avatar has earned more revenue in comparison to other films
